```python
import numpy as np
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import (
    OneHotEncoder,
    OrdinalEncoder,
    FunctionTransformer,
    MultiLabelBinarizer
)

Des_cat = ['Junior', 'Executive', 'Lead', 'Senior', 'Manager']
Edu_cat = ['High School', 'Bachelor', 'Master', 'PhD']

trf_2 = ColumnTransformer(
    transformers=[
        ('gender_binary',OrdinalEncoder(categories=[['Other','Female','Male']]),[13]),
        ('ord',OrdinalEncoder(categories=[Des_cat, Edu_cat]),[14, 18])
    ],
    remainder='passthrough'
)
```
